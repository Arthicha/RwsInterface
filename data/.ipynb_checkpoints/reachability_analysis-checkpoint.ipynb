# Reachability Analysis
### import python modules

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import sys
```

## <font color='red'>Section 1: Data Preprocessing </font> 

#### define column name

```python
name = ['id','X','Y','Z','r','p','y','Xo','Yo','Zo','ro','s1','s2','s3']
```

#### get data

```python
df1 = pd.read_csv('data1.csv').to_numpy()
df2 = pd.read_csv('data2.csv').to_numpy()
data = np.concatenate([df1,df2],axis=0)
data = pd.DataFrame(data[:,:-1],columns=name)
data
```

### combined data with same id

```python
def getUnique(cols,orien=None,score="average"):
    n = len(cols)
    scatter_datas = []
    features = data[cols].drop_duplicates().to_numpy()
    orientation = data[orien].drop_duplicates().to_numpy() if orien is not None else None
    no = 1 if orientation is None else orientation.shape[0]
    print(orientation)
    for o in range(no):
        scatter_data = [[] for i in range(n+1)]
        for fea in features:
            cond = True
            for i in range(n):
                cond = cond & (data[cols[i]] == fea[i])
            data_i = data.loc[cond]
            if orientation is not None:
                cond = True
                for i in range(n):
                    cond = cond & (data[orien[i]] == orientation[o][i])
                data_i = data_i.loc[cond]
            succ_rate = 0
            if score is "average":
                data_i_true = data_i.loc[(data['s1'] == 1) & (data['s2'] == 1)  & (data['s3'] == 1)]
                if data_i.shape[0] != 0:
                    succ_rate = data_i_true.shape[0]/data_i.shape[0]
                else:
                    scatter_data[-1].append(0.0)
            elif score is "maxz":
                Zos = data_i['Zo'].drop_duplicates().to_numpy()
                ii = 0
                for Zos_i in Zos:
                    ii += 1
                    data_i_true = data_i.loc[(data['s1'] == 1) & (data['s2'] == 1)  & (data['s3'] == 1) &
                                            (data['Zo'] == Zos_i)]
                    rate = data_i_true.shape[0]/data_i.loc[(data['Zo'] ==Zos_i)].shape[0]
                    succ_rate += np.power(0.515,ii)*rate
            elif score is "combined":
                Zos = data_i['Zo'].drop_duplicates().to_numpy()
                ii = 0
                for Zos_i in Zos:
                    ii += 1
                    data_i_true = data_i.loc[(data['s1'] == 1) & (data['s2'] == 1)  & (data['s3'] == 1) &
                                            (data['Zo'] == Zos_i)]
                    rate = data_i_true.shape[0]/data_i.loc[(data['Zo'] ==Zos_i)].shape[0]
                    succ_rate += np.power(0.515,ii)*rate
                data_i_true = data_i.loc[(data['s1'] == 1) & (data['s2'] == 1)  & (data['s3'] == 1)]
                if data_i.shape[0] != 0:
                    succ_rate *= data_i_true.shape[0]/data_i.shape[0]
                else:
                    succ_rate = 0.0
            scatter_data[-1].append(succ_rate)
            
            for i in range(n):
                scatter_data[i].append(fea[i])
            
        scatter_datas.append(scatter_data)
    return scatter_datas

def getMax(scatter_data):
    scatter_data = np.array(scatter_data)
    maxval = np.max(scatter_data[:,-1,:],axis=1)
    print(maxval.shape)
    max_loc = []
    for i in range(maxval.shape[0]):
        max_loc.append(scatter_data[i,:,scatter_data[i,-1,:] == maxval[i]][0])
    return max_loc
```

## <font color='red'>Section 2: Scatter Plot </font> 

### 2.1 using all data

```python
%matplotlib notebook
plt.rcParams["figure.figsize"] = (5,5)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
scatter_data = getUnique(["X","Y","Z"])
scatter_data = np.array(scatter_data[0])
ax.scatter(scatter_data[0],scatter_data[1],scatter_data[2], s=scatter_data[3]*300,c=scatter_data[3], cmap='plasma')
```

### 2.2 orientation-wise

#### orientation 1: upward

```python
scatter_data = getUnique(["X","Y","Z"],orien=["r","p","y"],score="combined")
scatter_data = np.array(scatter_data)
best_loc = getMax(scatter_data)
print(np.array(best_loc)[:,-1])
TITLE = ["upward","downward","left","right","front","back"]
n_sub = int(np.sqrt(scatter_data.shape[0]))
```

```python
%matplotlib notebook
plt.rcParams["figure.figsize"] = (9,6)
fig = plt.figure()
plt.subplots_adjust(hspace=0.5,wspace=0.5)
for i in range(scatter_data.shape[0]):
    sub_idx = n_sub*100+(int(scatter_data.shape[0]//n_sub)*10)+1+i
    ax = fig.add_subplot(sub_idx,projection='3d')
    ax.scatter(scatter_data[i][0],scatter_data[i][1],scatter_data[i][2], 
               s=np.power(scatter_data[i][-1],2)*50,c=scatter_data[i][-1], cmap='plasma')
    ax.scatter(best_loc[i][0],best_loc[i][1],best_loc[i][2], 
               s=50,c='green')
    x = best_loc[i][0]
    y = best_loc[i][1]
    z = best_loc[i][2]
    ax.plot3D([x,x],[y,y],[z,0],c='green',ls='--')
    ax.plot3D([-0.5,0.5],[y,y],[z,z],c='green',ls='--')
    ax.plot3D([x,x],[-0.5,0.5],[z,z],c='green',ls='--')
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.set_title(TITLE[i])
```

### plot one-by-one

```python
def plotHeatmap(scatter_data,i):
    global TITLE
    %matplotlib notebook
    plt.rcParams["figure.figsize"] = (6,6)
    fig = plt.figure()
    plt.tight_layout()
    ax = fig.add_subplot(111,projection='3d')
    ax.scatter(scatter_data[i][0],scatter_data[i][1],scatter_data[i][2], 
               s=np.power(scatter_data[i][-1],2)*300,c=scatter_data[i][-1], cmap='winter')
    ax.scatter(best_loc[i][0],best_loc[i][1],best_loc[i][2], 
               s=50,c='red')
    print(best_loc[i])
    x = best_loc[i][0]
    y = best_loc[i][1]
    z = best_loc[i][2]
    ax.plot3D([x,x],[y,y],[z,0],c='red',ls='-')
    ax.plot3D([-0.5,0.5],[y,y],[0,0],c='red',ls=':',lw=1)
    ax.plot3D([x,x],[-0.5,0.5],[0,0],c='red',ls=':',lw=1)
    ax.plot3D([x,x],[-0.5,y],[z,z],c='red',ls='-')
    ax.plot3D([-0.5,0.5],[-0.5,-0.5],[z,z],c='red',ls=':',lw=1)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.set_xlim([-0.5,0.5])
    ax.set_ylim([-0.5,0.5])
    ax.set_zlim([0.0,1.0])
    ax.set_title(TITLE[i]+" (score:"+str(int(10000*best_loc[i][-1])/10000)+")")
    ax.view_init(25, 45)
```

#### up (rpy = 0,0,0)

```python
plotHeatmap(scatter_data,0)
```

#### back (rpy = 0,0,90)

```python
plotHeatmap(scatter_data,1)
```

#### front (rpy = 0,0,-90)

```python
plotHeatmap(scatter_data,2)
```

#### down (rpy = 0,0,180)

```python
plotHeatmap(scatter_data,3)
```

#### right (rpy = 0,90,0)

```python
plotHeatmap(scatter_data,4)
```

#### left (rpy=0,-90,0)

```python
plotHeatmap(scatter_data,5)
```
